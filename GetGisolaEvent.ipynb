#GET EVENT DATA FROM GEOFON

```python
from obspy.clients.fdsn import Client
from obspy import UTCDateTime

# Initialize the client for GEOFON
client = Client("GEOFON")

# Define the time range for the events
start_time = UTCDateTime("2019-01-01")
end_time = UTCDateTime("2019-01-31")

# Define the latitude and longitude range for Indonesia
min_latitude = -11.0
max_latitude = 6.0
min_longitude = 95.0
max_longitude = 141.0

# Define the minimum magnitude
min_magnitude = 3.0

# Fetch events from GEOFON
catalog = client.get_events(starttime=start_time, endtime=end_time,
                            minlatitude=min_latitude, maxlatitude=max_latitude,
                            minlongitude=min_longitude, maxlongitude=max_longitude,
                            minmagnitude=min_magnitude, maxmagnitude=4.5)
```

```python
print(catalog)
```

```python
# Optionally, you can iterate over the events and print details
for event in catalog:
    origin = event.origins[0]
    magnitude = event.magnitudes[0]
    print(f"Event ID: {event.resource_id}")
    print(f"Time: {origin.time}")
    print(f"Latitude: {origin.latitude}, Longitude: {origin.longitude}")
    print(f"Depth: {origin.depth / 1000} km")  # depth is in meters, convert to km
    print(f"Magnitude: {magnitude.mag} {magnitude.magnitude_type}")
    print("-" * 40)
```

```python
from obspy.clients.fdsn import Client
from obspy import UTCDateTime

client = Client("geofon")

# # Define the event ID and retrieve the event
event_id = "gfz2019clna"
events = client.get_events(eventid=event_id)

# print(events)
```

```python
origin_time = events[0].origins[0].time
latitude = events[0].origins[0].latitude
longitude = events[0].origins[0].longitude

starttime = origin_time-60
endtime = starttime + 600  # Half hour after the origin time
```

```python
# Get the stations
inventory = client.get_stations(latitude=latitude, longitude=longitude, starttime=starttime, endtime=endtime, channel="BH?", level='response')

print(inventory)
```

```python
desired_stations = ["BKB", "BKNI", "BBJI", "BNDI", "CISI", "FAKI", "GENI", "GSI", "JAGI",
                    "LHMI", "LUWI", "MMRI", "MNAI", "PLAI", "PMBT", "PMBI", "SANI",
                    "SAUI", "SMRI", "SOEI", "TNTI", "TOLI", "TOLI2", "UGM", "YOGI"]
filtered_networks = []
for network in inventory:
    filtered_stations = [station for station in network.stations if station.code in desired_stations]
    if filtered_stations:
        new_network = network.copy()
        new_network.stations = filtered_stations
        filtered_networks.append(new_network)

filtered_inventory = inventory.copy()
filtered_inventory.networks = filtered_networks

# Print the filtered inventory
print(filtered_inventory)
```

```python
from obspy.clients.fdsn import Client
from obspy import UTCDateTime
from obspy.core.stream import Stream

# Initialize the client
client = Client("IRIS")

# Define the event ID and retrieve the event
event_id = "11176800"
events = client.get_events(eventid=event_id)

# Extract the event origin time and location
event = events[0]
origin = event.origins[0]
origin_time = origin.time
latitude = origin.latitude
longitude = origin.longitude

# Define the time window around the event origin time
starttime = origin_time - 5 * 60  # 5 minutes before
endtime = origin_time + 10 * 60  # 10 minutes after

# Define the radius to search for stations (in degrees)
radius = 10  # Adjust this value as needed

# Retrieve the station inventory
inventory = client.get_stations(latitude=latitude, longitude=longitude, maxradius=radius,
                                starttime=starttime, endtime=endtime, level="response")

# Initialize an empty stream to hold the waveforms
stream = Stream()

# Iterate over the stations in the inventory and query the waveform data
for network in inventory:
    for station in network:
        for channel in station:
            try:
                # Query the waveform data for the station
                waveform = client.get_waveforms(network.code, station.code, channel.location_code,
                                                channel.code, starttime, endtime)
                if waveform:
                    print(f"Station {network.code}.{station.code}.{channel.location_code}.{channel.code} detected the event.")
                    stream += waveform
            except Exception as e:
                print(f"Error retrieving waveform for station {network.code}.{station.code}.{channel.location_code}.{channel.code}: {e}")

# Process the collected waveforms as needed
if not stream:
    print("No stations detected the earthquake waveform.")
else:
    print(f"{len(stream)} traces found from {len(stream.select(network='CW'))} stations")
    # Save the collected waveforms to a file, if desired
    # stream.write("detected_waveforms.mseed", format="MSEED")

# Print summary of stations that detected the earthquake
for tr in stream:
    print(f"Detected by: {tr.id}")
```
